```python
file = open('dat1.txt', 'r+')

with open('dat1.txt', 'r') as f:
    data = f.readlines()

    for line in data:
        columns = line.split(",")
        print columns
        
from scipy.optimize import curve_fit        
import matplotlib.pyplot as plt
import numpy as np

datos = np.loadtxt('dat1.txt')
x=datos[:,0]
y=datos[:,1]
print x
print y

def func(x, A, B, C):
    return A * np.exp(-B * x) + C


Y = y + 0.2*np.random.normal(size=len(x))

popt, pcov = curve_fit(func, x, Y)

plt.figure()
plt.plot(x, Y, 'yo', label="Datos")
plt.plot(x, func(x, *popt), 'g-', label="Curva ajustada")

plt.grid()
plt.legend()
plt.title("Presion atmosferica vs. Altitud")
plt.xlabel("Altitud (Km)")
plt.ylabel("Presion (Mb)")

plt.show()
```
